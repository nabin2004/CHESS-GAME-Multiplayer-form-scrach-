## Start Latitude and Longitude

```python
df.Start_Lat
```

```python
sample = df.sample(int(0.1*len(df)))
sns.scatterplot(y=sample.Start_Lat, x=sample.Start_Lng,size=0.0001)
```

```python
import folium
```

```python
for index, row in df[['Start_Lat', 'Start_Lng']].sample(100).iterrows():
    lat,lon = row['Start_Lat'], row['Start_Lng']
    map = folium.Map(location=[lat, lng], zoom_start=12)
    marker = folium.Marker((lat, lng))
    marker.add_to(map)
map
    
```

```python
map = folium.Map(location=[lat, lng], zoom_start=12, markers="x")
marker = folium.Marker((lat, lng))
marker.add_to(map)
map
```

```python
from folium.plugins import HeatMap
```

```python
sample_df = df.sample(int(0.001 * len(df)))
lat_lon_pairs = list(zip(list(sample_df.Start_Lat), list(sample_df.Start_Lng)))
```

```python
map = folium.Map()
HeatMap(lat_lon_pairs).add_to(map)
map
```

On Sundays, the peak occurs between 10am and 3pm, unlike weekdays
